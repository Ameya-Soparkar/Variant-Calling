We need to create a function whcih gives us a dataframe for individual chromosome consisting of positions with multiple variation. Position with single variation will be dropped.

So the function works in this way:
1. The function takes two arguments 1 a dataframe and the chromosome number
2. Depending on the chromosome number the dataframe is filtered out in the next step and stored in 'x' variable.
3. in order to identify positions with multiple variation we have to find value counts of 'x', and create a new table using   reset index. (We have a table which has position number and its number of values)
4. This new table columns will be replaced as 'POS' and 'num, for our understanding  
5. We will filter out num greater than 1, since we want positions with multiple values, and create x_new table form it.
6. We will create a 'c' varable and convert the positions to numpy array and store in 'c' variable.
7. We will return a chrom dataframe(which will have all the positions which are in variable 'c') by using apply function in pandas.

```python
def get_chrom(df, chrom_no):
    chrom = df[df['CHROM'] == chrom_no] 
    x = chrom['POS'].value_counts()
    x = x.reset_index()
    x.columns= ['POS', 'num']
    x_new= x.loc[x['num'] > 1]
    c = x_new['POS'].to_numpy()
    chrom['POS'] = chrom['POS'].apply(lambda x : x if x in c else None)
    chrom.dropna(subset = ['POS'], inplace=True)
    return chrom
```

We will also create a function which will add a new column "CONFLICTING" which will have 0 and 1, 0 suggesting there is only one category in 'CLNSIG' column and 1 if there are more categories.

The function will be as follows:
1. The function will take one argument ie a dataframe.
1. For the dataframe we created it will add a column "CONFLICTING" with 0 value in it.
2. We will create a list of Position present in the dataframe, using set function we will remove the duplicates
3. We will loop over the position list we created, and filter each and every position individually and check for CLNSIG category
4. We will use set function and turn it into len, and if there are multiple categories the length will be more than 1.
5. If it is more than 1 then the CONFLICTING value will be converted to 1.


```python
def conflicting(c):
    c['CONFLICTING'] = 0
    pos_list = list(set(c.POS))
    for i in pos_list: 
        if len(set(c[c['POS'] == i]['CLNSIG'])) != 1:
            c.loc[c.POS == i, "CONFLICTING"] = 1      
        else:
            continue
    return c
```
